```python
# If your columns are named differently, e.g. 'Adj Close' vs 'Adjusted Close' vs Close or whatever the adjusted price column is called
spy_prices = spy_data['Close']  # Creates a new variable containing dates & closing prices

print(spy_data.columns)  #To check the column names
```

```python
spy_prices.head()
```

```python
# Extract adjusted close prices and combine
prices = pd.DataFrame({
    'SPY': spy_data['Close'],  # adjust column name if needed
    'QQQ': qqq_data['Close'],
    'IWM': iwm_data['Close']
}).dropna()

print("Combined prices shape:", prices.shape)
print("\nFirst 5 rows:")
prices.head()
```

```python
#Checks earliest date for each dataset
print("SPY start:", spy_data.index.min()) 
print("QQQ start:", qqq_data.index.min())
print("IWM start:", iwm_data.index.min())
```

```python
# Simple daily returns for SPY only
#pct.change calculates the percentage change bewteen consecutive days and dropna takes out the first date as no percentage increase - NoN
spy_returns = prices['SPY'].pct_change().dropna()
print("SPY daily returns shape:", spy_returns.shape)
print(spy_returns.head())
```

```python
# Log returns: ln(P_t / P_{t-1})
spy_log_returns = np.log(prices['SPY'] / prices['SPY'].shift(1)).dropna()
print("SPY log returns shape:", spy_log_returns.shape)
spy_log_returns.head()
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
spy_returns.plot(title="Simple Returns (SPY)", alpha=0.7)
plt.ylabel("Daily return")

plt.subplot(1, 2, 2)
spy_log_returns.plot(title="Log Returns (SPY)", alpha=0.7)
plt.ylabel("Daily log return")

plt.tight_layout()
plt.show()
```

```python
# Normalize everything to 100 at start
start_date = prices.index[0]
spy_normalized = prices['SPY'] / prices['SPY'].loc[start_date] * 100
cumulative_simple = (1 + spy_returns).cumprod() * 100
cumulative_log = np.exp(spy_log_returns.cumsum()) * 100

plt.figure(figsize=(12, 5))
plt.plot(spy_normalized.index, spy_normalized, label="SPY Price (norm to 100)", linewidth=2)
plt.plot(cumulative_simple.index, cumulative_simple, label="Cumulative Simple Returns", alpha=0.8)
plt.plot(cumulative_log.index, cumulative_log, label="Cumulative Log Returns (exp)", alpha=0.8)
plt.title("SPY: Price vs Cumulative Returns (normalized to 100)")
plt.ylabel("Index (100 = start)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
print("SPY Daily Log Returns Stats:")
print(spy_log_returns.describe() * 100)  # in percent

# Annualized
daily_mean = spy_log_returns.mean()
daily_vol = spy_log_returns.std()
annual_return = daily_mean * 252
annual_vol = daily_vol * np.sqrt(252)

print(f"\nAnnualized Return: {annual_return:.1%}")
print(f"Annualized Volatility: {annual_vol:.1%}")
print(f"Biggest daily gain: {spy_log_returns.max():.2%}")
print(f"Biggest daily loss: {spy_log_returns.min():.2%}")
```

```python
# Compute log returns for all ETFs
log_returns_all = np.log(prices / prices.shift(1)).dropna()

# Annualized stats table
annual_stats = pd.DataFrame({
    'Annual Return': log_returns_all.mean() * 252,
    'Annual Vol': log_returns_all.std() * np.sqrt(252)
}, index=['SPY', 'QQQ', 'IWM'])

print("Annualized Stats Comparison:")
print(annual_stats.round(3) * 100)  # in percent
```
