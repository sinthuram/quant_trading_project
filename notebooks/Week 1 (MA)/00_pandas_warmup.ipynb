```python
import pandas as pd
import numpy as np
```

```python
#Create's the dataset#
data = {
    "day": [1, 2, 3, 4, 5],
    "price": [100, 101, 102, 99, 103],
    "volume": [10_000, 11_000, 9_500, 12_000, 10_500],
}

df = pd.DataFrame(data)
df
```

```python
#Select one Column#
df["price"]
```

```python
#Create new column#
#pct.change - works out percentage change with the previous row#
df["return"] = df["price"].pct_change()
df
```

```python
#Filter rows giving an inequality#
df[df["price"]>100]
```

```python
#Calculates mean and standard deviation#
df["price"].mean(), df["price"].std()
```

```python
#Quick Plot for the above dataset#
import matplotlib.pyplot as plt

df["price"].plot(kind="line")
plt.title("Fake price series")
plt.show()

```
