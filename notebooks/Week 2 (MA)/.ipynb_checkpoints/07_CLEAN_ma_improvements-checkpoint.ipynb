Step 1 - Re-check your data loading, making sure notebook starts with clear loading code.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
spy = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/spy.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')   # change if needed
spy.columns = spy.columns.str.strip()

spy = pd.DataFrame({'Adj_Close': spy['Close']})

spy.sort_index(inplace=True)
spy.head()
```

```python
print(spy.columns)
```

Step 2 - Recreate the MA signal cleanly

```python
#MA Signal
short_window = 50
long_window = 200

spy["ma_short"] = spy["Adj_Close"].rolling(short_window).mean()
spy["ma_long"] = spy["Adj_Close"].rolling(long_window).mean()

spy["signal_raw"] = 0
spy.loc[spy["ma_short"] > spy["ma_long"], "signal_raw"] = 1
spy["signal"] = spy["signal_raw"].shift(1)

spy = spy.dropna(subset=["signal"])
spy[["Adj_Close", "ma_short", "ma_long", "signal_raw", "signal"]].tail()
```

"Dates" column ALWAYS needs to start with the oldest date on the top and the newest date on the bottom. For rolling mean to be calculated correctly and to stop look ahead bias.

Step 3 - Rebuild the transcation cost block

```python
#Returns and trade flag
spy["simple_return"] = spy["Adj_Close"].pct_change()
spy["trade"] = spy["signal"].diff().abs().fillna(0)
```

```python
transaction_cost = 0.001

spy["strategy_return_gross"] = spy["signal"] * spy["simple_return"]
spy["strategy_return_net"] = spy["strategy_return_gross"] - spy["trade"] * transaction_cost

spy[["simple_return", "signal", "trade", "strategy_return_gross", "strategy_return_net"]].head(15)
```

Transcation costs are important as they are one of the most important reasons why a good stratgy can become weak in practice.

Step 4 - Build equity curves again

```python
#Equity curves
spy["equity_bh"] = (1 + spy["simple_return"]).cumprod()
spy["equity_gross"] = (1 + spy["strategy_return_gross"]).cumprod()
spy["equity_net"] = (1 + spy["strategy_return_net"]).cumprod()
```

```python
#Plot equity curves
plt.figure(figsize=(12,6))
plt.plot(spy.index, spy["equity_bh"], label="Buy and Hold")
plt.plot(spy.index, spy["equity_gross"], label="MA Gross")
plt.plot(spy.index, spy["equity_net"], label="MA Net")
plt.title("Equity Curves")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

#Costs are not making a huge difference in MA Gross and MA Net - key to see if costs are eating the edge 
```

Step 5 - Recreate the metrics table

```python
#Metric functions
def annualized_return(r):
    return (1 + r.mean())**252 - 1

def annualized_vol(r):
    return r.std() * np.sqrt(252)

def sharpe_ratio(r):
    vol = annualized_vol(r)
    return np.nan if vol == 0 else annualized_return(r) / vol

def max_drawdown(equity):
    return (equity / equity.cummax() - 1).min()
```

```python
#Metrics table
metrics = pd.DataFrame({
    "Buy & Hold": [
        annualized_return(spy["simple_return"]),
        annualized_vol(spy["simple_return"]),
        sharpe_ratio(spy["simple_return"]),
        max_drawdown(spy["equity_bh"])
    ],
    "MA Gross": [
        annualized_return(spy["strategy_return_gross"]),
        annualized_vol(spy["strategy_return_gross"]),
        sharpe_ratio(spy["strategy_return_gross"]),
        max_drawdown(spy["equity_gross"])
    ],
    "MA Net": [
        annualized_return(spy["strategy_return_net"]),
        annualized_vol(spy["strategy_return_net"]),
        sharpe_ratio(spy["strategy_return_net"]),
        max_drawdown(spy["equity_net"])
    ]
}, index=["Annual Return", "Annual Volatility", "Sharpe", "Max Drawdown"])

metrics.round(4)
```

Step 6 - Rebuild parameter sensitivity more neatly

```python
#Creating a parameter list
param_pairs = [(20, 100), (50, 200), (100, 300), (30, 150)]
results = []
```

```python
#Loop through combinations
for s, l in param_pairs:
    temp = spy.copy()
    temp["ma_short"] = temp["Adj_Close"].rolling(s).mean()
    temp["ma_long"] = temp["Adj_Close"].rolling(l).mean()
    temp["signal_raw"] = (temp["ma_short"] > temp["ma_long"]).astype(int)
    temp["signal"] = temp["signal_raw"].shift(1)
    temp["simple_return"] = temp["Adj_Close"].pct_change()
    temp["trade"] = temp["signal"].diff().abs().fillna(0)
    temp["strategy_return_net"] = temp["signal"] * temp["simple_return"] - temp["trade"] * transaction_cost
    temp = temp.dropna(subset=["strategy_return_net"])
    
    eq = (1 + temp["strategy_return_net"]).cumprod()
    
    results.append({
        "short_window": s,
        "long_window": l,
        "cum_return": eq.iloc[-1] - 1,
        "sharpe": sharpe_ratio(temp["strategy_return_net"]),
        "max_drawdown": max_drawdown(eq)
    })

sensitivity_df = pd.DataFrame(results)
sensitivity_df
```

Step 7 - Save the sensitivity charts

```python
#Bar chart for cumulative return
plt.figure(figsize=(10,5))
labels = sensitivity_df.apply(lambda x: f'{int(x["short_window"])}/{int(x["long_window"])}', axis=1)
plt.bar(labels, sensitivity_df["cum_return"])
plt.title("Cumulative Return by MA Parameters")
plt.ylabel("Cumulative Return")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

```python
#Bar chart for Sharpe
plt.figure(figsize=(10,5))
plt.bar(labels, sensitivity_df["sharpe"])
plt.title("Sharpe Ratio by MA Parameters")
plt.ylabel("Sharpe")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

Step 8 - Recreate the drawdown check

```python
equity_net = spy["equity_net"]
max_dd_net = (equity_net / equity_net.cummax() - 1).min()
print(f"Net strategy max drawdown: {max_dd_net:.2%}")
```

## Reflections on MA strategy improvements

- Transcation cost assumption used: 0.10%
- Costs changed performance materially, even testing with a 1% showed a drastics drop in profits.
- Best MA pair in small test: 100/300
- Results were robust across the selction of results (only slighlty dropping in sharpe ratio for some pairs)
- Gross results are useful, but net results are more realistic
