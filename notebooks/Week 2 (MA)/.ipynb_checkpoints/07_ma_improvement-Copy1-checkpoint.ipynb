```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Step 1 - Load SPY and rebuild the baseline MA strategy

```python
# Load your 3 WSJ CSV files from data folder
# Replace 'SPY.csv', 'QQQ.csv', 'IWM.csv' with your ACTUAL filenames
spy_data = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/spy.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')
qqq_data = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/qqq.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')
iwm_data = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/iwm.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')

#Clean up column names for each data set (they have unneccasary space at the beginning of the column
spy_data.columns = spy_data.columns.str.strip()
qqq_data.columns = qqq_data.columns.str.strip()
iwm_data.columns = iwm_data.columns.str.strip()
```

```python
# Extract adjusted close prices and combine
spy = pd.DataFrame({'Adj_Close': spy_data['Close']  })

#Making sure dates start with oldest to newest
spy.sort_index(inplace=True)

# Simple and log returns
spy["simple_return"] = spy["Adj_Close"].pct_change()
spy["log_return"] = np.log(spy["Adj_Close"] / spy["Adj_Close"].shift(1))
spy = spy.dropna(subset=["simple_return", "log_return"])
spy.head()
```

```python
#MA Signal
short_window = 50
long_window = 200

spy["ma_short"] = spy["Adj_Close"].rolling(short_window).mean()
spy["ma_long"] = spy["Adj_Close"].rolling(long_window).mean()

spy["signal_raw"] = 0
spy.loc[spy["ma_short"] > spy["ma_long"], "signal_raw"] = 1
spy["signal"] = spy["signal_raw"].shift(1)

spy = spy.dropna(subset=["signal"])
spy[["Adj_Close", "ma_short", "ma_long", "signal_raw", "signal"]].tail()
```

Step 2 - Add transaction costs

- We'll assume a small cost per time you change position. This is a rough but important first realism check.

```python
#Identify trade days
spy["trade"] = spy["signal"].diff().abs()
spy["trade"] = spy["trade"].fillna(0)
spy[["signal_raw", "signal", "trade"]].head(970)

#If the signal changes from 0 to 1 or 1 to 0, trade becomes 1
```

```python
#Defining a transcation cost and choosing a simple cost assumption
transaction_cost = 0.001  # 0.10%
```

```python
# Strategy returns before/ after cost
spy["strategy_return_gross"] = spy["signal"] * spy["simple_return"]

# Subtract cost only on days when a trade occurs
spy["strategy_return_net"] = spy["strategy_return_gross"] - spy["trade"] * transaction_cost

spy[["simple_return", "signal", "trade", "strategy_return_gross", "strategy_return_net"]].head(970)
```

Step 3 - Compare gross vs net performance

```python
#Equity curves
spy["equity_gross"] = (1 + spy["strategy_return_gross"]).cumprod()
spy["equity_net"] = (1 + spy["strategy_return_net"]).cumprod()
spy["equity_bh"] = (1 + spy["simple_return"]).cumprod()

plt.figure(figsize=(12, 5))
plt.plot(spy.index, spy["equity_bh"], label="Buy & Hold", linewidth=1.5)
plt.plot(spy.index, spy["equity_gross"], label="MA Gross", linewidth=1.5)
plt.plot(spy.index, spy["equity_net"], label="MA Net of Costs", linewidth=1.5)
plt.title("Equity Curves: Buy & Hold vs MA (Gross vs Net)")
plt.ylabel("Wealth (start = 1)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#Summary metrics
def perf_metrics(ret_series):
    equity = (1 + ret_series).cumprod()
    ann_ret = ret_series.mean() * 252
    ann_vol = ret_series.std() * np.sqrt(252)
    sharpe = ann_ret / ann_vol if ann_vol != 0 else np.nan
    max_dd = (equity / equity.cummax() - 1).min()
    cum_ret = equity.iloc[-1] - 1
    return pd.Series({
        "cumulative_return": cum_ret,
        "annual_return": ann_ret,
        "annual_vol": ann_vol,
        "sharpe_like": sharpe,
        "max_drawdown": max_dd
    })

summary = pd.DataFrame({
    "Buy & Hold": perf_metrics(spy["simple_return"]),
    "MA Gross": perf_metrics(spy["strategy_return_gross"]),
    "MA Net": perf_metrics(spy["strategy_return_net"])
}).T

summary
```

Step 4 - Parameter sensitivity test

Now testing to see whether the result depends too much on the exact MA windows. Testing difference combination of MA windows will show variation in stats.

```python
def run_ma_strategy(df, short_window, long_window, cost=0.001):
    temp = df.copy()
    temp["ma_short"] = temp["Adj_Close"].rolling(short_window).mean()
    temp["ma_long"] = temp["Adj_Close"].rolling(long_window).mean()
    temp["signal_raw"] = (temp["ma_short"] > temp["ma_long"]).astype(int)
    temp["signal"] = temp["signal_raw"].shift(1)
    temp = temp.dropna(subset=["signal"])
    temp["trade"] = temp["signal"].diff().abs().fillna(0)
    temp["ret_gross"] = temp["signal"] * temp["simple_return"]
    temp["ret_net"] = temp["ret_gross"] - temp["trade"] * cost
    equity = (1 + temp["ret_net"]).cumprod()
    return {
        "short": short_window,
        "long": long_window,
        "cum_return": equity.iloc[-1] - 1,
        "sharpe": (temp["ret_net"].mean() * 252) / (temp["ret_net"].std() * np.sqrt(252)) if temp["ret_net"].std() != 0 else np.nan,
        "max_dd": (equity / equity.cummax() - 1).min()
    }
```

```python
#Trying a small set of MA pairs, to see the cumulative returns for differnet windows
results = []
for s, l in [(20, 100), (50, 200), (30, 150), (100, 300)]:
    results.append(run_ma_strategy(spy[["Adj_Close", "simple_return"]], s, l, cost=transaction_cost))

results_df = pd.DataFrame(results)
results_df
```

Step 5 - Visualise the sensitivity results

```python
#Bar chart
plt.figure(figsize=(10, 4))
plt.bar(results_df["short"].astype(str) + "/" + results_df["long"].astype(str), results_df["cum_return"])
plt.title("Cumulative Return for Different MA Parameters")
plt.ylabel("Cumulative Return")
plt.grid(True, axis="y", alpha=0.3)
plt.show()
```

```python
#Optional Sharpe comparision
plt.figure(figsize=(10, 4))
plt.bar(results_df["short"].astype(str) + "/" + results_df["long"].astype(str), results_df["sharpe"])
plt.title("Sharpe-like Ratio for Different MA Parameters")
plt.ylabel("Sharpe-like")
plt.grid(True, axis="y", alpha=0.3)
plt.show()
```

Step 6 - Add a drawdown rule check

Not changing the strategy yet, just checking what happens if you stop trading after a bad period.

```python
#Max drawdown of net strategy
equity_net = spy["equity_net"]
max_dd_net = (equity_net / equity_net.cummax() - 1).min()
print(f"Net strategy max drawdown: {max_dd_net:.2%}")
```

That drawdown is large - a decent strategy needs to be tweaked.

## MA strategy improvements – reflections

- Transaction cost assumption used:
  - 0.10 %

- Did costs materially change performance?
  - No, all main stats stayed the same and performance would of produced results.

- Which MA pair looked best in my small test?
  - 100/300 (ever so slightly difference)

- Did the performance look robust or overly dependent on the exact parameters?
  - Looks robust, only a selection of MA windows that show it is not successful.

- What I learned from comparing gross vs net results:
  - Gross results show an accurate view of the effects of costs. Net results show the realistic outcome.

Using a Heatmap to show the robustness of the MA windows, by testing several variations

```python
results = []

short_windows = range(10, 60, 5)     # e.g. 10,15,...,55
long_windows = range(80, 300, 20)    # e.g. 80,100,...,280

for s in short_windows:
    for l in long_windows:
        if s < l:  # IMPORTANT: short must be less than long
            res = run_ma_strategy(
                spy[["Adj_Close", "simple_return"]],
                s, l,
                cost=transaction_cost
            )
            results.append(res)

results_df = pd.DataFrame(results)
```

```python
heatmap_data = results_df.pivot(
    index="short",
    columns="long",
    values="sharpe"   # or "cum_return"
)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(heatmap_data, aspect='auto')

plt.colorbar(label="Sharpe Ratio")

plt.xticks(
    ticks=range(len(heatmap_data.columns)),
    labels=heatmap_data.columns,
    rotation=45
)

plt.yticks(
    ticks=range(len(heatmap_data.index)),
    labels=heatmap_data.index
)

plt.xlabel("Long MA Window")
plt.ylabel("Short MA Window")
plt.title("MA Crossover Strategy Sharpe Ratio Heatmap")

plt.show()
```
