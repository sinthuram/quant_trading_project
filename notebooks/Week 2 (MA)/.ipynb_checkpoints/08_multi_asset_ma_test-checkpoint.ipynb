Step 2 - Define your strategy function

Instead of repeating code three times, make one reusuable function for transcation costs. Reusuable function makes your research cleaner and lets you compare assets consistently.

```python
transaction_cost = 0.001

def ma_strategy_backtest(df, short_window=50, long_window=200, cost=0.001):
    temp = df.copy()
    temp["simple_return"] = temp["Adj_Close"].pct_change()
    temp["ma_short"] = temp["Adj_Close"].rolling(short_window).mean()
    temp["ma_long"] = temp["Adj_Close"].rolling(long_window).mean()
    temp["signal_raw"] = (temp["ma_short"] > temp["ma_long"]).astype(int)
    temp["signal"] = temp["signal_raw"].shift(1)
    temp["trade"] = temp["signal"].diff().abs().fillna(0)
    temp["strategy_return_gross"] = temp["signal"] * temp["simple_return"]
    temp["strategy_return_net"] = temp["strategy_return_gross"] - temp["trade"] * cost
    temp = temp.dropna(subset=["strategy_return_net"])
    temp["equity_bh"] = (1 + temp["simple_return"]).cumprod()
    temp["equity_gross"] = (1 + temp["strategy_return_gross"]).cumprod()
    temp["equity_net"] = (1 + temp["strategy_return_net"]).cumprod()
    return temp
```

Step 3 - Define metrics function

```python
def annualized_return(r):
    return (1 + r.mean())**252 - 1

def annualized_vol(r):
    return r.std() * np.sqrt(252)

def sharpe_ratio(r):
    vol = annualized_vol(r)
    return np.nan if vol == 0 else annualized_return(r) / vol

def max_drawdown(equity):
    return (equity / equity.cummax() - 1).min()
```

Step 4 - Run the strategy on all ETFs

Had a bug due to incorrect coding. When trying to build results to run a strategy for all ETFs, it was not possible, due to files not having a ticker which is include in results. 

```python
# --- Run strategy across all ETFs ---

results = []

for ticker, df in data.items():          # ← iterate over data, not files
    bt = ma_strategy_backtest(df, short_window=50, long_window=200, cost=transaction_cost)
    
    results.append({
        "Ticker": ticker,
        "Annual Return BH":  annualized_return(bt["simple_return"]),
        "Annual Vol BH":     annualized_vol(bt["simple_return"]),
        "Sharpe BH":         sharpe_ratio(bt["simple_return"]),
        "Max DD BH":         max_drawdown(bt["equity_bh"]),
        "Annual Return Net": annualized_return(bt["strategy_return_net"]),
        "Annual Vol Net":    annualized_vol(bt["strategy_return_net"]),
        "Sharpe Net":        sharpe_ratio(bt["strategy_return_net"]),
        "Max DD Net":        max_drawdown(bt["equity_net"]),
        "Final Equity BH":   bt["equity_bh"].iloc[-1],
        "Final Equity Net":  bt["equity_net"].iloc[-1]
    })

results_df = pd.DataFrame(results)
results_df.round(4)
```

Step 6 - Parameter grid for Sharpe

Only evaluating combinations where the short window is smaller than the long window.

```python
#Choosing a grid
short_windows = [20, 30, 40, 50, 60, 70, 80, 90, 100]
long_windows = [100, 120, 140, 160, 180, 200, 220, 240, 260]
```

Build Sharpe matrix for one ETF

```python
base_df = data["SPY"]
grid = pd.DataFrame(index=short_windows, columns=long_windows, dtype=float)

for s in short_windows:
    for l in long_windows:
        if s >= l:
            grid.loc[s, l] = np.nan
        else:
            bt = ma_strategy_backtest(base_df, short_window=s, long_window=l, cost=transaction_cost)
            grid.loc[s, l] = sharpe_ratio(bt["strategy_return_net"])

grid
```

Step 7 - Plot the heat map for Sharpe Ratio

```python
plt.figure(figsize=(8,6))
sns.heatmap(grid, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Sharpe Ratio Heatmap – SPY MA Strategy")
plt.xlabel("Long Window")
plt.ylabel("Short Window")
plt.show()
```

Step 8 - Heatmap for Cumulative return

```python
ret_grid = pd.DataFrame(index=short_windows, columns=long_windows, dtype=float)

for s in short_windows:
    for l in long_windows:
        if s >= l:
            ret_grid.loc[s, l] = np.nan
        else:
            bt = ma_strategy_backtest(base_df, short_window=s, long_window=l, cost=transaction_cost)
            ret_grid.loc[s, l] = bt["equity_net"].iloc[-1] - 1

plt.figure(figsize=(8,6))
sns.heatmap(ret_grid, annot=True, fmt=".2f", cmap="viridis")
plt.title("Cumulative Return Heatmap – SPY MA Strategy")
plt.xlabel("Long Window")
plt.ylabel("Short Window")
plt.show()
```
