- Testing whether my strategy returns are statistically convincing
- We are not proving the strategy works forever, just checking whether the average return looks meaningfully different from zero.
- Checking this holds up compared with buy and hold

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline

#scipy.stats for the t-test
```

Step 3 - Load the data

```python
spy = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/spy.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')   # change if needed
spy.columns = spy.columns.str.strip()

spy = pd.DataFrame({'Adj_Close': spy['Close']})

spy.sort_index(inplace=True)
spy.head()
```

```python
short_window = 50
long_window = 200
transaction_cost = 0.001

spy["simple_return"] = spy["Adj_Close"].pct_change()
spy["ma_short"] = spy["Adj_Close"].rolling(short_window).mean()
spy["ma_long"] = spy["Adj_Close"].rolling(long_window).mean()
spy["signal_raw"] = (spy["ma_short"] > spy["ma_long"]).astype(int)
spy["signal"] = spy["signal_raw"].shift(1)
spy["trade"] = spy["signal"].diff().abs().fillna(0)
spy["strategy_return_net"] = spy["signal"] * spy["simple_return"] - spy["trade"] * transaction_cost

spy = spy.dropna(subset=["strategy_return_net"])
```

Step 4 - Define the hypotheses

## Hypothesis test

Null hypothesis: The mean daily strategy return is equall to zero.

Alternative hypothesis: The mean daily strategy return is not equal to zero.

It does NOT prove the strategy is great, instead tells whether the average daily return is distinguishable from zero in this sample.

Step 5 - Run the one-sample t-test

```python
strategy_r = spy["strategy_return_net"].dropna()

t_stat, p_value = stats.ttest_1samp(strategy_r, 0)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
```

- The t-statistic shows how far the sample mean is from zero realtive to volatility.
- The p-value tells you how suprising that result is under the null hypothesis

Step 6 - Interpret the test

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: the mean strategy return is significantly different from zero.")
else:
    print("Fail to reject the null hypothesis: there is not strong evidence that the mean strategy return differs from zero.")
```

Step 7 - Test buy and hold method as well (This gives a useful comparision)

```python
bh_r = spy["simple_return"].dropna()
t_stat_bh, p_value_bh = stats.ttest_1samp(bh_r, 0)

print(f"Buy & Hold T-statistic: {t_stat_bh:.4f}")
print(f"Buy & Hold P-value: {p_value_bh:.4f}")
```

Step 8 - Compare mean returns directly

```python
comparison = pd.DataFrame({
    "Buy & Hold": [
        bh_r.mean(),
        bh_r.std(),
        len(bh_r)
    ],
    "MA Net": [
        strategy_r.mean(),
        strategy_r.std(),
        len(strategy_r)
    ]
}, index=["Mean Daily Return", "Daily Volatility", "Observations"])

comparison
```

Is the mean daily return positive?
- Yes for both
  
Is it larger than the noise?
- Yes, as we have larger than 3 for t-test 

Step 9 - Confidence interval

```python
mean_r = strategy_r.mean()
std_r = strategy_r.std()
n = len(strategy_r)

se = std_r / np.sqrt(n)
ci_low = mean_r - 1.96 * se
ci_high = mean_r + 1.96 * se

print(f"Approx 95% CI for mean daily return: [{ci_low:.6f}, {ci_high:.6f}]")
```

Interpretation
- If the interval includes zero, that weakens the case for a real positive edge.
- In this case, the interval does not include zero.

## Interpretation

Was the strategy mean return significantly different from zero?
- As the strategy p_value < 0.05, therefore the mean return is statistically different from zero
- If p_value < 0.05, you typically say the mean return is statistically different from zero.
- If p_value >= 0.05, you do not have strong evidence it differs from zero.
