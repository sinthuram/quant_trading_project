```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Step 1 - Load data and standarise the price columns

```python
xlu = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/xlu.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')   # change if needed

xlu.columns = xlu.columns.str.strip()

xlu = pd.DataFrame({'Adj_Close': xlu['Close']})

xlu.sort_index(inplace=True)
xlu.head()
```

```python
xlp = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/xlp.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')   # change if needed

xlp.columns = xlp.columns.str.strip()

xlp = pd.DataFrame({'Adj_Close': xlp['Close']})

xlp.sort_index(inplace=True)
xlp.head()
```

```python
prices = pd.DataFrame({
    "XLU": xlu["Adj_Close"],
    "XLP": xlp["Adj_Close"]
}).dropna()

prices.head()
```

Step 2 - Plot both assets together

```python
plt.figure(figsize=(12,6))
plt.plot(prices.index, prices["XLU"], label="XLU")
plt.plot(prices.index, prices["XLP"], label="XLP")
plt.title("XLU vs XLP Prices")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Step 3 - Compute the ratio

```python
prices["ratio"] = prices["XLU"] / prices["XLP"]

plt.figure(figsize=(12,5))
plt.plot(prices.index, prices["ratio"], label="XLU / XLP")
plt.title("Price Ratio")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Step 5 - Add rolling mean and rolling std

```python
lookback = 60
prices["ratio_mean"] = prices["ratio"].rolling(lookback).mean()
prices["ratio_std"] = prices["ratio"].rolling(lookback).std()

plt.figure(figsize=(12,5))
plt.plot(prices.index, prices["ratio"], label="Ratio")
plt.plot(prices.index, prices["ratio_mean"], label=f"{lookback}-day Mean")
plt.title("Ratio and Rolling Mean")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Step 6 - Build the z-score

```python
prices["zscore"] = (prices["ratio"] - prices["ratio_mean"]) / prices["ratio_std"]

plt.figure(figsize=(12,5))
plt.plot(prices.index, prices["zscore"], label="Z-score")
plt.axhline(0, color="black", linewidth=1)
plt.axhline(1, color="red", linestyle="--", linewidth=1)
plt.axhline(-1, color="green", linestyle="--", linewidth=1)
plt.axhline(2, color="darkred", linestyle="--", linewidth=1)
plt.axhline(-2, color="darkgreen", linestyle="--", linewidth=1)
plt.title("XLU/XLP Z-score")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### How to interpret it

- Z-score near 0: ratio near recent average.

- Z-score above +2 or below -2: unusually stretched relationship.

- Your future trades will likely be based on those extremes.

# Proposed trading logic

The strategy is based on the z-score of the spread between XLU and XLP.
The z-score measures how many standard deviations the current spread
is away from its historical average.

**Core assumption:** when the spread diverges significantly from its mean,
it is expected to revert back toward zero over time.

### Signal rules

- When z-score > +2: XLU is unusually expensive relative to XLP.
  → Short the spread (sell XLU, buy XLP)

- When z-score < -2: XLU is unusually cheap relative to XLP.
  → Long the spread (buy XLU, sell XLP)

### Exit rules

- Exit the position when the z-score returns to within ±0.5 of zero.
  → The spread has normalised, close both legs and take profit.

### Stop-loss rule

- If z-score moves beyond ±3, exit immediately.
  → The spread is moving further against us; structural change may be
  occurring and the mean reversion assumption may no longer hold.

### Look-ahead bias control

- All signals are shifted forward by one day before backtesting.
  → A signal generated on day T is only acted on at day T+1 open,
  reflecting realistic execution timing.

### Thresholds summary

| Signal         | Z-score condition | Action                  |
|----------------|-------------------|-------------------------|
| Long spread    | z < -2            | Buy XLU, sell XLP       |
| Short spread   | z > +2            | Sell XLU, buy XLP       |
| Exit           | -0.5 < z < +0.5   | Close both positions    |
| Stop-loss      | z > +3 or z < -3  | Exit immediately        |

Step 7 - Correlation check

```python
#Correlation of returns
returns = prices[["XLU", "XLP"]].pct_change().dropna()
corr = returns.corr().iloc[0,1]
print(f"Return correlation: {corr:.3f}")
```

Correlation alone does NOT prove a pair trade works, but it helps you understand whether the assets move similarly on a broad level.
