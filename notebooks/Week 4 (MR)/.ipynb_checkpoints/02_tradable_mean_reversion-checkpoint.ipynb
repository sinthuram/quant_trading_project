```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Step 1 - Load data and price dataframe

```python
xlu = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/xlu.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')   # change if needed
xlp = pd.read_csv('/Users/Sinthu/Desktop/(Quant Trading Project)/data/xlp.csv', parse_dates=['Date'], date_format='%m/%d/%y', index_col='Date')   # change if needed

xlu.columns = xlu.columns.str.strip()
xlp.columns = xlp.columns.str.strip()

xlu = pd.DataFrame({'Adj_Close': xlu['Close']})
xlp = pd.DataFrame({'Adj_Close': xlp['Close']})

xlu.sort_index(inplace=True)
xlp.sort_index(inplace=True)
```

```python
prices = pd.DataFrame({
    "XLU": xlu["Adj_Close"],
    "XLP": xlp["Adj_Close"]
}).dropna()

prices.head()
```

```python
prices["ratio"] = prices["XLU"] / prices["XLP"]

lookback = 60
prices["ratio_mean"] = prices["ratio"].rolling(lookback).mean()
prices["ratio_std"] = prices["ratio"].rolling(lookback).std()

prices["zscore"] = (prices["ratio"] - prices["ratio_mean"]) / prices["ratio_std"]
```

```python
prices.head()
prices[["XLU", "XLP", "ratio", "ratio_mean", "ratio_std", "zscore"]].head(120)
```

## Step 2 - Define the entry and exit rules

## A clean rule set is:

- Enter short spread when z-score is above +2.
- Enter long spread when z-score is below -2.
- Exit when z-score comes back near 0, or within a small band such as −0.5 to 0.5

## Trading rules

- If the ratio is too high, XLU may be relatively expensive versus XLP.
- If the ratio is too low, XLU may be relatively cheap versus XLP.
- The strategy expects the ratio to move back toward its mean.

Step 3 - Build a basic signal column

```python
#Raw signal

entry_threshold = 2
exit_threshold = 0.5

prices["signal_raw"] = 0
prices.loc[prices["zscore"] < -entry_threshold, "signal_raw"] = 1
prices.loc[prices["zscore"] > entry_threshold, "signal_raw"] = -1

prices["signal_raw"].value_counts(dropna=False)
```

Step 4 - Add a holding state

A raw signal alone is not enough. You need a position that stays open until the exit condition is met.

This step allows the strategy to behave like a real trade than a one-day indicator.

```python
#Position logic

prices["position"] = 0

for i in range(1, len(prices)):
    prev_pos = prices["position"].iloc[i-1]
    z = prices["zscore"].iloc[i]
    raw = prices["signal_raw"].iloc[i]
    
    if prev_pos == 0:
        prices.iloc[i, prices.columns.get_loc("position")] = raw
    else:
        if abs(z) <= exit_threshold:
            prices.iloc[i, prices.columns.get_loc("position")] = 0
        else:
            prices.iloc[i, prices.columns.get_loc("position")] = prev_pos

prices[["zscore", "signal_raw", "position"]].tail(70)
```

Step 5 - Shift the position

To avoid look-ahead bias, shift the position by one day before calculating returns.

This matters as you can NOT trade on information from the same 'CLOSE' unless your execution assumption supports that. Shifting keeps the backtest more honest.

```python
prices["position_shifted"] = prices["position"].shift(1)
prices[["position", "position_shifted"]].head(70)
```

Step 6 - Plot the signal

```python
# Z-score with thresholds

plt.figure(figsize=(12,5))
plt.plot(prices.index, prices["zscore"], label="Z-score")
plt.axhline(entry_threshold, color="red", linestyle="--", linewidth=1)
plt.axhline(-entry_threshold, color="green", linestyle="--", linewidth=1)
plt.axhline(exit_threshold, color="gray", linestyle=":", linewidth=1)
plt.axhline(-exit_threshold, color="gray", linestyle=":", linewidth=1)
plt.axhline(0, color="black", linewidth=1)
plt.title("XLU/XLP Z-score with Entry and Exit Bands")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
#Position plot

plt.figure(figsize=(12,4))
plt.plot(prices.index, prices["position"], label="Position", drawstyle="steps-post")
plt.title("Strategy Position Over Time")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

Step 7 - Add a basic trade count

```python
#Count trades

prices["trade"] = prices["position"].diff().abs().fillna(0)
trade_count = (prices["trade"] > 0).sum()
print(f"Trade count: {trade_count}")
```

Important to look at the trade count as the mean-reversion strategy can look good but still be too active once costs are applied.
