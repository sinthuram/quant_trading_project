## Strategy descripiton

- This is mean reversion strategy pairs trade, between the ETF: XLU and XLP.
- The spread is represented by the price ratio and z-score.
- The strategy enters when the spread becomes stretched - this refers to the current price being far away from the actual/ expected value which it will revert to.
- The strategy enters when the spread comes back towards normal - this is when the price comes back to expected value.
- Positions are shifted by one day to reduce look-ahead bias.

## Results summary

- Gross performance was "0.0718".
- Net performance was "0.0614" after costs.
- Performance is annual return.
- The strategy traded 260 times.
- Max drawdown was "-0.2231"
- The best parameter region was "90 - lookback, 2.5 - entry and 0.5 - exit".
- The heat maps showed hot spots, not an overall warm area. Showing it works for a narrow region.
- Costs hurt the edge drastically and this was showing in the plots for gross and net.
- The pair did not give high annual returns - which was bad.
- It gave a low volitality - which was good.

## Robustness judgement

Did the strategy work across a range of parameters?
- Yes, only in one narrow setting.
- This seems like a positive however, is a bad sign.
- The good results are not broad enough to call this robust.
- Too sensitive to thresholds and lookback choice - regardless of this strategy showing some mean - reversion behaviour.
   
Did nearby parameter choices behave similarly?
- No, heatmap shows a big drop in paramter choices nearby.

Did transaction costs change the conclusion?
- Yes, drastic change in results from gross to net.

Would I trust this strategy live today?
- No, this is a good learning project.
